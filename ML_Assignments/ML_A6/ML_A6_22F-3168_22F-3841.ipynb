```python
import idx2numpy
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python

# Load and preprocess data
data_train_imgs = idx2numpy.convert_from_file('archive-2/train-images.idx3-ubyte')
data_train_lbls = idx2numpy.convert_from_file('archive-2/train-labels.idx1-ubyte')

data_test_imgs = idx2numpy.convert_from_file('archive-2/t10k-images.idx3-ubyte')
data_test_lbls = idx2numpy.convert_from_file('archive-2/t10k-labels.idx1-ubyte')

# Reshape and normalize images
data_train_imgs = data_train_imgs.reshape((60000, 28, 28, 1)).astype('float32') / 255
data_test_imgs = data_test_imgs.reshape((10000, 28, 28, 1)).astype('float32') / 255

# Convert labels to categorical format
data_train_lbls = tf.keras.utils.to_categorical(data_train_lbls, 10)
data_test_lbls = tf.keras.utils.to_categorical(data_test_lbls, 10)

# Define AlexNet-inspired model
model_alexnet = models.Sequential([
    layers.Conv2D(96, kernel_size=(3, 3), strides=1, activation='relu', input_shape=(28, 28, 1), padding='same'),
    layers.MaxPooling2D(pool_size=(2, 2), strides=2),

    layers.Conv2D(256, kernel_size=(3, 3), strides=1, activation='relu', padding='same'),
    layers.MaxPooling2D(pool_size=(2, 2), strides=2),

    layers.Conv2D(384, kernel_size=(3, 3), strides=1, activation='relu', padding='same'),
    layers.Conv2D(384, kernel_size=(3, 3), strides=1, activation='relu', padding='same'),

    layers.Conv2D(256, kernel_size=(3, 3), strides=1, activation='relu', padding='same'),
    layers.MaxPooling2D(pool_size=(2, 2), strides=2),

    layers.Flatten(),
    layers.Dense(4096, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(4096, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])

# Compile the model
model_alexnet.compile(optimizer='adam',
                      loss='categorical_crossentropy',
                      metrics=['accuracy'])

# Train the model
model_alexnet.fit(data_train_imgs, data_train_lbls, epochs=3, batch_size=32, validation_split=0.2)

# Evaluate the model
loss_alexnet, acc_alexnet = model_alexnet.evaluate(data_test_imgs, data_test_lbls)
print(f"Modified AlexNet test accuracy: {acc_alexnet}")
```
